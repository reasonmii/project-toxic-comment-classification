### Most of the text length are within 500 characters, with some up to 5,000 characters long.

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r"what's", "what is ", text)
    text = re.sub(r"\'s", " ", text)
    text = re.sub(r"\'ve", " have ", text)
    text = re.sub(r"can't", "cannot ", text)
    text = re.sub(r"n't", " not ", text)
    text = re.sub(r"i'm", "i am ", text)
    text = re.sub(r"\'re", " are ", text)
    text = re.sub(r"\'d", " would ", text)
    text = re.sub(r"\'ll", " will ", text)
    text = re.sub(r"\'scuse", " excuse ", text)
    text = re.sub('\W', ' ', text)
    text = re.sub('\s+', ' ', text)
    text = text.strip(' ')
    return text
```

```python
train['comment_text'] = train['comment_text'].map(lambda com : clean_text(com))
test['comment_text'] = test['comment_text'].map(lambda com : clean_text(com))
```

## What are the most common words in toxic comments?

```python
from wordcloud import WordCloud, STOPWORDS
from collections import Counter
```

```python
non_toxic_df = train[train['non-toxic']==True]
non_toxic_word_counter = Counter()
for comment in non_toxic_df['comment_text']:
    words = comment.split(' ')
    for word in words:
        if word not in STOPWORDS:
            non_toxic_word_counter[word] += 1

wc = WordCloud(background_color="white", max_words=1000)
wc.generate_from_frequencies(non_toxic_word_counter)
fig, ax = plt.subplots(figsize=(12, 12))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
toxic_df = train[train['non-toxic']==False]
toxic_df['comment_text']
```

```python
toxic_word_counter = Counter()
for comment in toxic_df['comment_text']:
    words = comment.split(' ')
    for word in words:
        if word not in STOPWORDS:
            toxic_word_counter[word] += 1
```

```python
wc = WordCloud(background_color="white", max_words=1000)
wc.generate_from_frequencies(toxic_word_counter)
fig, ax = plt.subplots(figsize=(12, 12))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
toxic_word_counter.most_common(20)
```

## Seems that some words occur a lot often both in toxic / non-toxic comments.

```python
toxic_words = {word for (word, freq) in toxic_word_counter.most_common(300)}
non_toxic_words = {word for (word, freq) in non_toxic_word_counter.most_common(300)}
toxic_words -= non_toxic_words
len(toxic_words)
```

```python
toxic_words
```

```python
train['number_of_toxic_words'] = 0
for i in range(len(train)):
    comment = train.iloc[i]['comment_text']
    cnt = 0
    words = comment.split(' ')
    for word in words:
        if word in toxic_words:
            cnt += 1
    train.at[i, 'number_of_toxic_words'] = cnt
```

```python
def build_classifier_model():
    text_input = tf.keras.layers.Input(shape=(), dtype=tf.string, name='input1')
    preprocessing_layer = hub.KerasLayer(tfhub_handle_preprocess, name='preprocessing')
    encoder_inputs = preprocessing_layer(text_input)
    encoder = hub.KerasLayer(tfhub_handle_encoder, trainable=True, name='BERT_encoder')
    outputs = encoder(encoder_inputs)
    net = outputs['pooled_output']
    net = tf.keras.layers.Dropout(0.2)(net)
    side_input = tf.keras.layers.Input(shape=(2, ), dtype=tf.float32, name='input2')
    net = tf.keras.layers.Concatenate(axis=-1)([net, side_input])
    net = tf.keras.layers.Dense(1, activation=None, name='classifier')(net)
    return tf.keras.Model(inputs=[text_input, side_input], outputs=net)
```
